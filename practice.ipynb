```python
from typing import List
```

```python
def minProduct(arr: List[int]) -> int:
    output = arr[0]
    left, right = 0, len(arr)-1
    while left <= right:
        if arr[left] < arr[right]:
            output = min(output, arr[left])
            break
        middle = left + right // 2
        output = min(output, arr[middle])
        if arr[middle] > arr[left]:
            left = middle + 1
        else:
            right = middle - 1
    return output

minProduct([4, 5, 6, 7, 0, 1, 2])
```

```python
def max_product_subarr(arr: List[int]) -> int:
    output = max(arr)
    curMin, curMax = 1, 1
    for i in arr:
        if i == 0:
            curMax, curMin = 1, 1
        temp = curMax * i
        curMax = max(i*curMax, i*curMin, i)
        curMin = min(temp, i*curMin, i)
        output = max(output, curMax)
    return output

max_product_subarr([2, 3, -2, 4])
```
