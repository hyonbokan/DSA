```python
def maxArea(arr: List[int]) -> int:
    left, right = 0, len(arr) - 1
    output = 0
    while left < right:
        area = (right - left) * max(arr[left], arr[right])
        output = max(output, area)
        if arr[left] < arr[right]:
            left += 1
        else:
            right += 1
    return output
maxArea([2,3,5,8,1,3,4,5,9])
```
