```python
def product_arr_noSelf(arr: List[int]) -> List[int]:
    output = [1] * len(arr)
    prefix = 1
    for i in range(len(arr)):
        output[i] = prefix
        prefix *= arr[i]
    postfix = 1
    for i in range(len(arr) -1, -1, -1):
        output[i] *= postfix
        postfix *= arr[i]
    return output
product_arr_noSelf([4, 5, 6, 7])
```

```python
def search_rotate_arr(arr: List[int], target: int) -> int:
    left, right = 0, len(arr)-1
    while left <= right:
        middle = (left + right) // 2
        if target == arr[middle]:
            return arr[middle]
        if arr[middle] > arr[left]:
            if target < arr[left] or target > arr[middle]:
                left = middle + 1
            else: 
                right = middle - 1
        else:
            if target > arr[right] or target < arr[middle]:
                right = middle - 1
            else:
                left = middle + 1
    return -1

search_rotate_arr([4,5,6,7,1,2,3], 5)
```
