```python
def product_arr_noSelf(arr: List[int]) -> List[int]:
    output = [1] * len(arr)
    prefix = 1
    for n in range(len(arr)):
        output[n] = prefix
        prefix *= arr[n]
    postfix = 1
    for n in range(len(arr)-1, -1, -1):
        output[n] *= postfix
        postfix *= arr[n]
    return output
product_arr_noSelf([4, 5, 6, 7])
```
